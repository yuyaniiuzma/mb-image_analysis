```python
import pandas as pd
```

```python
df = pd.read_csv('tracking_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python

df_rev = df[['Matching_ID', 'Glass_Type(1:Outer 2:Inner 3:Pair)', 'MB_Count','MB_Sheet_Count','MB_LD_Vison_X_Pos', 'MB_LD_Vison_Y_Pos', 'MB_LD_Vison_C_Pos', 'MB_LD_Vison_Score', 'MB_PUT_Vison_X_Pos', 'MB_PUT_Vison_Y_Pos','MB_PUT_Vison_C_Pos', 'MB_Push_Force']]
```

```python
df_rev.info()
```

```python
df_rev = df_rev[(df_rev['MB_Sheet_Count'] == 7) & (df_rev['Glass_Type(1:Outer 2:Inner 3:Pair)'] == 1)]
```

```python
df_rev.head()
```

```python
df_rev.reset_index(inplace=True)
```

```python
df_rev.head()
```

```python
bins_X = [-2900, -2500, -2100, -1700, -1300, -900]
bins_Y = [-1400, -1100, -700, -400, -100, 200, 500, 800, 1200]
labels_X = [1,2,3,4,5]
labels_Y = [8,7,6,5,4,3,2,1]
df_rev['MB_LD_Vison_X'] = pd.cut(df_rev['MB_LD_Vison_X_Pos'], bins=bins_X, labels=labels_X)
df_rev['MB_LD_Vison_Y'] = pd.cut(df_rev['MB_LD_Vison_Y_Pos'], bins=bins_Y, labels=labels_Y)
df_rev.head()
```

```python
df_rev['MB_LD_Vison_X'].value_counts()
df_rev['MB_LD_Vison_Y'].value_counts()
```

```python
df_G = df_rev.sort_values(by = ['MB_LD_Vison_X', 'MB_LD_Vison_Y'])
df_G = df_G[['Matching_ID', 'MB_LD_Vison_X', 'MB_LD_Vison_Y', 'Glass_Type(1:Outer 2:Inner 3:Pair)', 'MB_Count','MB_Sheet_Count','MB_LD_Vison_X_Pos', 'MB_LD_Vison_Y_Pos', 'MB_LD_Vison_C_Pos', 'MB_LD_Vison_Score', 'MB_PUT_Vison_X_Pos', 'MB_PUT_Vison_Y_Pos','MB_PUT_Vison_C_Pos', 'MB_Push_Force']]
df_G.reset_index(inplace=True)
df_G
```

```python
df_out = pd.read_csv('output.csv')
df_rev2 = pd.concat([df_G, df_out],axis=1)
df_rev2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(2,2,figsize=(8,5),tight_layout=True)
# plt.figure(figsize=(8,6))
sns.regplot(data=df_rev2, x= 'mean', y='MB_LD_Vison_Score', ax=axes[0,0])
sns.regplot(data=df_rev2, x= 'std', y='MB_LD_Vison_Score', ax=axes[0,1])
# sns.regplot(data=df_rev2, x= 'kurt', y='MB_LD_Vison_Score', ax=axes[1,0])
# sns.regplot(data=df_rev2, x= 'skewness', y='MB_LD_Vison_Score', ax=axes[1,1])
sns.regplot(data=df_rev2, x= 'peak_distance', y='MB_LD_Vison_Score', ax=axes[1,0])

axes[0,0].set_ylim(60,95)
axes[0,1].set_ylim(60,95)
# axes[1,0].set_ylim(65,95)
# axes[1,1].set_ylim(65,95)
axes[1,0].set_ylim(60,95)
# plt.ylim(65,95)
plt.show()
```
